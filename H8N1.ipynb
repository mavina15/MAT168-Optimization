```python
# import libraries
import gurobipy as gp
from gurobipy import GRB
```

```python
# create model
model = gp.Model()

# create decision variables
x1 = model.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x1")
x2 = model.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x2")
x3 = model.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x3")
x4 = model.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x4")

# set objective function
model.setObjective(1*x1 + 2*x2 + 1*x3 + 1*x4, GRB.MAXIMIZE)

# add constraints
model.addConstr(2*x1 + 1*x2 + 5*x3 + 1*x4 <= 8, "c1")
model.addConstr(2*x1 + 2*x2 + 4*x4 <= 12, "c2")
model.addConstr(3*x1 + 1*x2 + 2*x3 <= 18, "c3")

# optimize model
model.optimize()

# print optimal solution and optimal value
print("\n")
print("Optimal Solution: x1={}, x2={}, x3={}, x4={}".format(x1.x, x2.x, x3.x, x4.x))
print("Optimal Value: {}".format(model.objVal))
```
